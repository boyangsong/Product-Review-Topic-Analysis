```python
#Baidu API test
from aip import AipNlp
import csv
import pandas as pd
from pandas.core.frame import DataFrame

""" 你的 APPID AK SK """
# Use Baidu API for sentiment analysis
APP_ID = '11479249'
API_KEY = 'Qdl4zUz7ksZvQiooffkwWzoY'
SECRET_KEY = '6d8RQ5p6Vj972zBKDdLygomo1cQ7ihcn'
client = AipNlp(APP_ID, API_KEY, SECRET_KEY)

# Load the document
def output():
    urls = []
    with open('c:/Users/Administrator/Desktop/bose.csv', "r") as f:
        reader = csv.reader(f)
        for row in reader:
            urls.append(row[1])
    return urls

# Perform sentiment analysis and load the result to a Json file and a new dictionary
def sentimentClassify():
    x = output()
    i=1
    pp=[]
    np=[]
    con=[]
    sen=[]
    all={}
    for i in range(933):
        text=x[i]
        result=client.sentimentClassify(text);
        import time
        time.sleep(0.25)
        print(result)
        if "error_code" in result.keys():
            pp.append(" ")
            np.append(" ")
            con.append(" ")
            sen.append(" ")
            all['positive_prob'] = pp
            all['negative_prob'] = np
            all['confidence'] = con
            all['sentiment'] = sen
        else:
            data = result['items']
            items = data[0]
            positive_prob = items['positive_prob']
            pp.append(positive_prob)
            negative_prob = items['negative_prob']
            np.append(negative_prob)
            confidence = items['confidence']
            con.append(confidence)
            sentiment = items['sentiment']
            sen.append(sentiment)
            all['positive_prob'] = pp
            all['negative_prob'] = np
            all['confidence'] = con
            all['sentiment'] = sen
    return all
# Conver dictionary to CSV
def add(ulist):
    csv_input = pd.read_csv('c:/Users/Administrator/Desktop/bose.csv',encoding='gbk')
    pp = DataFrame(ulist['positive_prob'])
    csv_input["positive_prob"] = pp
    csv_input.to_csv('c:/Users/Administrator/Desktop/bose.csv', index=False,encoding='gbk')
    np = DataFrame(ulist['negative_prob'])
    csv_input["negative_prob"] = np
    csv_input.to_csv('c:/Users/Administrator/Desktop/bose.csv', index=False, encoding='gbk')
    con = DataFrame(ulist['confidence'])
    csv_input["confidence"] = con
    csv_input.to_csv('c:/Users/Administrator/Desktop/bose.csv', index=False, encoding='gbk')
    sen = DataFrame(ulist['sentiment'])
    csv_input["sentiment"] = sen
    csv_input.to_csv('c:/Users/Administrator/Desktop/bose.csv', index=False, encoding='gbk')

if __name__ == '__main__':
    ALL=sentimentClassify()
    add(ALL)
```

```python
#LDA model，based on gensim：
#-*- coding: utf-8 -*-
import pandas as pd
import 1121warnings
warnings.filterwarnings(action='ignore',category=UserWarning,module='gensim')

#Initializing
negfile = 'c:/Users/Administrator/Desktop/negfile_cut1.txt'
posfile = 'c:/Users/Administrator/Desktop/posfile_cut1.txt'
stoplist = 'c:/Users/Administrator/Desktop/stoplist.txt'

neg = pd.read_csv(negfile, encoding = 'utf-8', header = None, engine='python') #读入数据
pos = pd.read_csv(posfile, encoding = 'utf-8', header = None, engine='python')
stop = pd.read_csv(stoplist, encoding = 'utf-8', header = None, sep = 'tipdm', engine='python')

stop = [' ', ''] + list(stop[0])

neg[1] = neg[0].apply(lambda s: s.split(' ')) 
neg[2] = neg[1].apply(lambda x: [i for i in x if i not in stop]) 
pos[1] = pos[0].apply(lambda s: s.split(' '))
pos[2] = pos[1].apply(lambda x: [i for i in x if i not in stop])

from gensim import corpora, models

#Negative sentiment topic analysis
neg_dict = corpora.Dictionary(neg[2]) 
neg_corpus = [neg_dict.doc2bow(i) for i in neg[2]]
neg_lda = models.LdaModel(neg_corpus, num_topics = 3, id2word = neg_dict) 
for i in range(3):
    w=neg_lda.print_topic(i) 
    print(w)
#Positive sentiment topic analysis
pos_dict = corpora.Dictionary(pos[2])
pos_corpus = [pos_dict.doc2bow(i) for i in pos[2]]
pos_lda = models.LdaModel(pos_corpus, num_topics = 3, id2word = pos_dict)
for i in range(3):
    z=pos_lda.print_topic(i) 
    print(z)
```
